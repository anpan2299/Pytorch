```python
#file_path = input("local do arquivo")
file_path = 'Licitações - 1ª Remessa  -  12042023  - INTEL SEMICONDUTORES DO BRASIL - CÓPIA IV- REN.txt' 
with open('Servidor/' + file_path, 'r') as f:
    lines = f.readlines()
```

```python
lines
```

```python
licita = True
for i, line in enumerate(lines):
    if line == 'Registro(s) \n' and licita:
        start = i
        licita = False
    if line == 'Retificações \n':
        end = i
file = lines[start:end] 
print(len(file))
```

```python
file_clean = []
for x in file:
    clean_element = re.sub(r"[ +\n]", ' ', x)
    if clean_element.strip():
        file_clean.append(clean_element)
print(len(file_clean))
file_clean
```

```python
licitacoes = []
licitacao = {}
for i,x in enumerate(file_clean):
    if x == 'Código ' or x == 'Edital: ' or x =='Data de Entrega  ' or x == 'Modalidade ' or x=='Licitante ':
        licitacao[x] = file_clean[i+1]
    if x == 'Descrição Objeto ' or x == 'Objeto ':
        start = i+1
        j = 1
        while(file_clean[start + j] != 'Licitante '):
            j += 1
        end = start + j
        spell = TextBlob(''.join(file_clean[start:end]).lower())
        licitacao['Descrição Objeto'] = str((spell.translate(from_lang='pt',to='es')).translate(from_lang='es',to='pt'))
        
            
        
    if 'Licitante ' in licitacao:
        if ("servidor" in licitacao['Descrição Objeto'] or "servidores" in licitacao['Descrição Objeto'] or "Servidor" in licitacao['Descrição Objeto'] or "Servidores" in licitacao['Descrição Objeto']):
            licitacoes.append(licitacao)
            licitacao = {}
        else:
            licitacao = {}
    
    
```
